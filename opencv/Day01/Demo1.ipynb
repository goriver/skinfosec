```python
img = cv2.imread('./DATA/00-puppy.jpg')
print(img)
```

* 원래는 rgb인데
* opencv에서만큼은 **BGR** 을 사용한다.

```python
img_bgr = cv2.imread('./DATA/00-puppy.jpg')
plt.imshow(img_bgr)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# 가지고 있는 입력 데이터를 convert하는 함수
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('./DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap='gray')
# cmap은 gray로 바꿔서 출력하라는 것이다.
```

## resize images

```python
img_rgb.shape
```

```python
img = cv2.resize(img_rgb, (1300, 275)) # 세로가 1950 -> 275로 변경됨
plt.imshow(img)
```

## by ration

```python
w_ratio = 0.5
h_ratio = 0.5
new_img = cv2.resize(img_rgb, (0,0), img, w_ratio, h_ratio)


plt.imshow(img)
```

```python
## flipping images
```

```python
new_img = cv2.flip(new_img, 0)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(new_img, 1) # 1 좌우대칭
plt.imshow(new_img)
```

```python
new_img = cv2.flip(new_img, -1) # -1은 원래대로
plt.imshow(new_img)
```

## 이미지 저장하기

```python
cv2.imwrite('my_new_puppy.jpg', new_img)
```
